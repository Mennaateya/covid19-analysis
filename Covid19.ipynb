```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset.
#Dataset url ==> https://covid.ourworldindata.org/data/owid-covid-data.csv
df = pd.read_csv('owid-covid-data.csv')
```

```python
pd.set_option('future.no_silent_downcasting', True)
#No cases reported so fill with 0
df['total_cases']=df['total_cases'].fillna(0)
#Fill missing deaths with the average
df['total_deaths']=df['total_deaths'].fillna(df['total_deaths'].mean())
```

```python
#No new cases reported so fill with 0
df['new_cases'] = df['new_cases'].fillna(0)
#No new deaths reported, so fill with 0
df['new_deaths'] = df['new_deaths'].fillna(0)
```

```python
#These columns are the average of new cases and deaths over 7 days to make the data smoother
#Fill missing cases with the average
df['new_cases_smoothed'] = df['new_cases_smoothed'].fillna(df['new_cases_smoothed'].mean())
#Fill missing deaths with the average
df['new_deaths_smoothed'] = df['new_deaths_smoothed'].fillna(df['new_deaths_smoothed'].mean())
```

```python
#Filling columns that contains text
df['iso_code'] = df['iso_code'].fillna("Unknown")
df['continent'] = df['continent'].fillna("Unknown")
df['location'] = df['location'].fillna("Unknown")
```

```python
df.isna().sum()
```

```python
#Load Egypt data
df_egy=df[df['location']=='Egypt']
print(df_egy.tail())
```

```python
#Average confirmed cases and deaths in Egypt
print("Average confirmed cases in Egypt: ", round(df_egy['total_cases'].mean(), 2))
print("Average deaths in Egypt: ", round(df_egy['total_deaths'].mean(), 2))

#Highest and lowest confirmed cases in Egypt
print("Highest confirmed cases in Egypt: ", df_egy['total_cases'].max())
print("Lowest confirmed cases in Egypt: ", df_egy['total_cases'].min())

#Average daily new cases and deaths in Egypt
print("Average daily new cases in Egypt: ", round(df_egy['new_cases'].mean(), 2))
print("Average daily new deaths in Egypt: ", round(df_egy['new_deaths'].mean(), 2))
```

```python
#Data Type Conversion
df['date']=pd.to_datetime(df['date'])
print(df['date'].dtype)
```

```python
#Filtering data
df_egypt = df[df['location'] == 'Egypt']
start_date = '2021-01-01'
end_date = '2021-6-12'
data_filtered = df[(df['date'] >= start_date) & (df['date'] <= end_date)]
```

```python
#Feature Engineering: Adding new columns
df['death_rate'] = df['total_deaths'] / df['total_cases']
df['new_cases_rate'] = df['new_cases'] / df['total_cases']
```

```python
#Plot total COVID-19 cases and deaths in Egypt over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='total_cases', data=df_egy, label='Total Cases')
sns.lineplot(x='date', y='total_deaths', data=df_egy, label='Total Deaths')
plt.title('COVID-19 Cases and Deaths in Egypt Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
#Plot daily new COVID-19 cases in Egypt
plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='new_cases', data=df_egy)
plt.title('Daily New COVID-19 Cases in Egypt')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.show()
```

```python
#Compare total COVID-19 cases in Egypt, Italy, and the United States
countries = ['Egypt', 'Italy', 'United States']
df_countries = df[df['location'].isin(countries)]

plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='total_cases', hue='location', data=df_countries)
plt.title('Total COVID-19 Cases in Different Countries')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend(title='Country')
plt.show()
```

```python
#Plot COVID-19 death rate in Egypt over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='death_rate', data=df_egy)
plt.title('COVID-19 Death Rate in Egypt Over Time')
plt.xlabel('Date')
plt.ylabel('Death Rate')
plt.show()
```
